```python
#This file is intended to generate suitable training data, train a network and then analyse the network output.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import csv



```

```python
b = np.arange(0.2, 3.2, 0.01)
d = np.arange(0.1, 1.0, 0.01)

inp2, inp1 = np.meshgrid(b, d)
out = np.sqrt( 1 + (2*inp1*inp2)**2 ) / np.sqrt( (1-inp2**2)**2 + (2*inp1*inp2)**2)

plt.contourf(inp2, inp1, out)
plt.colorbar()
plt.xlabel('b')
plt.ylabel('d')
plt.show()
```

```python
inpPairs = np.array([np.reshape(inp1, inp1.size), np.reshape(inp2, inp2.size)])

inpPairs = inpPairs.transpose()

outPairs = np.reshape(out, out.size)
print(inpPairs.shape)
np.savetxt("inputs.csv", inpPairs, delimiter=",", fmt='%.10f')
np.savetxt("outputs.csv", outPairs, delimiter=",", fmt='%.10f')



```
